## Perturbation size analysis

```python
from matplotlib.ticker import PercentFormatter
eps = 0.031
p_pgd_nopnml_eps_08 = pathlib.Path(r"./results/cifar/code_v2/pgd_madry_diff_eps/cifar_adversarial_results_20200513_213908")
p_pgd_nopnml_eps_05 = pathlib.Path(r"./results/cifar/code_v2/pgd_madry_diff_eps/cifar_adversarial_results_20200513_213625")
p_pgd_nopnml_eps_04 = pathlib.Path(r"./results/cifar/code_v2/pgd_madry_diff_eps/cifar_adversarial_results_20200513_213341")
p_pgd_nopnml_eps_03 = pathlib.Path(r"./results/cifar/code_v2/pgd_madry_diff_eps/cifar_adversarial_results_20200513_161318")
p_pgd_nopnml_eps_03_200_iter = pathlib.Path(r"./results/cifar/code_v2/pgd_200iter_nopnml")
p_natural_nopnml = pathlib.Path(r".\results\cifar\code_v2\cifar_natural_nopnml")
adv1, params = analyze_utilities_v2.load_exp_result_from_dir(p_pgd_nopnml_eps_03_200_iter, merge_splitted_res_flag=False)
natural, params = analyze_utilities_v2.load_exp_result_from_dir(p_natural_nopnml)
diff = adv1.adversarial_sample - natural.adversarial_sample
plt.hist(diff.flatten()/eps, bins=20, weights=np.ones(len(diff.flatten())) / len(diff.flatten()))
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
# plt.xticks(np.arange(-2, 1, step=1))  # Set label locations.
plt.xticks(np.arange(-1,2), ['$-\epsilon$', '0','$\epsilon$'])
plt.ylabel('Density')
plt.xlabel('Perturbation size')

plt.savefig(os.path.join(output_path, 'fig_cifar_pert_size_hist.jpeg'),
            dpi=200, bbox_inches=plt.tight_layout()) if is_save_fig else None

per = (diff.flatten().abs() < diff.max()-0.001).sum().item()/diff.flatten().__len__()
print("Percentage of perturbations that are not maximal: {}%".format(100*per))
```

```python
# Change the perturbation to be only +-\epsilon:

# plt.hist(a.flatten(), bins=20)
import numpy as np
cond_pos = np.logical_and((diff < diff.max()).numpy(), (diff>=0).numpy())
adv1.adversarial_sample[cond_pos] = natural.adversarial_sample[cond_pos] + diff.max()
cond_neg = np.logical_and((diff > diff.min()).numpy(), (diff<0).numpy())
adv1.adversarial_sample[cond_neg] = natural.adversarial_sample[cond_neg] + diff.min()
new_diff = adv1.adversarial_sample - natural.adversarial_sample
plt.hist(new_diff.flatten(), bins=20)
adv1.dump(p_pgd_nopnml_eps_03_300_iter, "maximize_pert.t")  # Results of eval: Accuracy: 0.6295, Loss: 1.02933271484375 Vs Accuracy: 0.6266, Loss: 1.03365380859375
# pNML results: Accuracy: 0.6695, Loss: 1.03286796875
```

```python
adv1, params = analyze_utilities_v2.load_exp_result_from_dir(p_pgd_nopnml_eps_03_300_iter)
diff = adv1.adversarial_sample - natural.adversarial_sample
import numpy as np
ther = 0.001
cond_pos = np.logical_and((diff < diff.max()-ther).numpy(), (diff>=0).numpy())
adv1.adversarial_sample[cond_pos] = natural.adversarial_sample[cond_pos]
cond_neg = np.logical_and((diff > diff.min()+ther).numpy(), (diff<0).numpy())
adv1.adversarial_sample[cond_neg] = natural.adversarial_sample[cond_neg]
new_diff = adv1.adversarial_sample - natural.adversarial_sample
plt.hist(new_diff.flatten(), bins=20)
adv1.dump(p_pgd_nopnml_eps_03_300_iter, "zero_non_max_pert.t")  # Results: Accuracy: 0.6299, Loss: 1.02778837890625
# pNML Results: Accuracy: 0.6721, Loss: 1.0283724609375
```
