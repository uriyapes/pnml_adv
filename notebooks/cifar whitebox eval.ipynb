# 2. Results

### 2.1 Accuracy Vs. Adaptive-attack & PGD-attack strength

#### Load pgd_nopnml_diff_eps

```python
indices = [i for i in range(0, 9999+1, 1)]
subdir_list = analyze_utilities_v2.get_subdir_list(pgd_nopnml_diff_eps_path)
print(subdir_list)
df_pgd_nopnml = analyze_utilities_v2.results_dirs_to_df(subdir_list)
df_pgd_nopnml = df_pgd_nopnml.sort_values("epsilon", ignore_index=True)
```

#### Load pgd_pnml_diff_eps

```python
subdir_list = analyze_utilities_v2.get_subdir_list(pgd_pnml_diff_eps_path)
df_pgd_pnml = analyze_utilities_v2.results_dirs_to_df(subdir_list, indices).sort_values("epsilon", ignore_index=True)
```

#### Load adaptive_diff_eps

```python
subdir_list = analyze_utilities_v2.get_subdir_list(adaptive_diff_eps_path)
print(subdir_list)
df_adaptive = analyze_utilities_v2.results_dirs_to_df(subdir_list)
df_adaptive = df_adaptive.sort_values("epsilon", ignore_index=True)
```

### Plot Accuracy Vs \epsilon

```python
plt.figure(0)
# plt.plot(df_adaptive['epsilon'], df_adaptive['acc'], 'go-', label='Adaptive our scheme') # Adaptive our scheme
plt.plot(df_pgd_nopnml.loc[:, 'epsilon'], df_pgd_nopnml.loc[:, 'acc'], 'ro--', label='Carmon et al. [2019]')
plt.plot(df_pgd_pnml.loc[:, 'epsilon'], df_pgd_pnml.loc[:, 'acc'], 'bo-', label='Ours') # PGD our scheme

axes = plt.gca()
axes.set_ylim([0.0, 0.9])
plt.axvline(x=0.031, color='k', linestyle='dotted', label='$\epsilon=.031$')
plt.legend(fontsize=15, ncol=1, loc=1)
plt.xlabel('Attack strength')
plt.ylabel('Accuracy')
plt.savefig(os.path.join(output_path, 'fig_cifar_best_vs_attack_strength.pdf'), dpi=200,
            bbox_inches=plt.tight_layout()) if is_save_fig else None
plt.show()
print("df_pgd_nopnml:\n" + str(df_pgd_nopnml))
print("df_pgd_pnml:\n" + str(df_pgd_pnml))
# print("df_adaptive:\n" + str(df_adaptive))
```

```python
indices = [i for i in range(0, 9999+1, 1)]
subdir_list = analyze_utilities_v2.get_subdir_list(fgsm_nopnml_diff_eps_path)
print(subdir_list)
df_fgsm_nopnml = analyze_utilities_v2.results_dirs_to_df(subdir_list)
df_fgsm_nopnml = df_fgsm_nopnml.sort_values("epsilon", ignore_index=True)

subdir_list = analyze_utilities_v2.get_subdir_list(fgsm_pnml_diff_eps_path)
print(subdir_list)
df_fgsm_pnml = analyze_utilities_v2.results_dirs_to_df(subdir_list)
df_fgsm_pnml = df_fgsm_pnml.sort_values("epsilon", ignore_index=True)
```
